### General Analysis

#### Mean height of cities above mean sea level
* Colombo- 1m
* Matara- 2m
* Nuwara Eliya- 1,868m
* Kandy- 500m
* Monaragala-151m
* Kurunegala-116m
* Jaffna- 5m

```python
#let us make a plot of the altitudes of each city
plt.figure(figsize=(10,6))
cities = ['Colombo','Matara','Nuwara Eliya','Kandy','Monaragala','Kurunegala','Jaffna']
altitude = [1,2,1868,500,151,116,5]
sns.barplot(x=altitude, y=cities, color='purple')
plt.title("Altitude of cities above mean sea level")
plt.ylabel("Cities")
plt.xlabel("Altitude above mean sea level (meters)")
plt.show()
```

```python
altitude = [1,2,1868,500,151,116,5]
mean_hcho = [0.000165, 0.000104, 0.000104, 0.000122, 0.000137, 0.00014, 0.000111]
cities = ['Colombo', 'Matara', 'Nuwara Eliya', 'Kandy', 'Monaragala', 'Kurunegala', 'Jaffna']

# plotting a line plot
plt.figure(figsize=(10,6))
plt.plot(altitude, mean_hcho, marker='o', color='blue', linestyle='-')
# marking the name of the cities on the plot
for i, city in enumerate(cities):
    plt.annotate(city, (altitude[i], mean_hcho[i]), textcoords='offset points', xytext=(0,10), ha='center')
plt.title('Altitude vs Mean HCHO Concentration')
plt.xlabel('Altitude(m)')
plt.ylabel('Mean HCHO Concentration')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
altitude_correlation_coeff = np.corrcoef(altitude, mean_hcho)
print('Correlation coefficient between altitude and mean HCHO concentration:', altitude_correlation_coeff[0,1])
```
